Nie widać znaczącej korelacji pomiędzy liczbą pustych miejsc w sudoku, a liczbą zamian potrzebną do jego rozwiązania. Można jednak zauważyć, że ilość zamian jest zdecydowanie większa dla przykładu most_difficult. Wynika to prawdopodobnie z faktu, że sudoku to zostało specjalnie zaprojektowane, by być trudne do rozwiązania przez komputer. Pozostałe sudoku to zwykłe sudoku stworzone do rozwiązywania przez ludzi, jak widać w ich przypadku niezależnie od stopnia trudności algorytm potrzebował około 100 000 zamian.

Liczba możliwych stanów dla przykładu most_difficult wynosi: $ (6!)^7\cdot7!\cdot9! \approx 1.83 \cdot 10^{29} $, czyli dużo więcej niż liczba potrzebnych zamian do jego rozwiązania.
